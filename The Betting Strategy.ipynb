# The Betting Strategy

You're given 100 dollars and you joined a betting game. You will give your bet to the banker and the banker will flip a coin. If the coin lands on heads, you will get your bet money back. Otherwise, you will lose your bet money and then bet for another round. The challenge is to hold the money until 25 or more rounds. If you're suggested to give 1 dollar as an initial bet and increase the bet by 1 dollar every time you get tails, would you take the suggestion? (Remember, it's not about getting money more than 100 dollars! It's about holding the money as long as possible)

```python
#Required modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define coinflip function
def coinflip():
    return(np.random.randint(0,2))
```

Example: <br>
**Round 1 = Heads** <br>
Your money will stay 100, your bet amount will stay 1<br>
**Round 2 = Tails**<br>
Your money will be 99, your bet amount will be 2<br>
**Round 3 = Tails**<br>
Your money will be 97, your bet amount will be 3<br>
**Round 4 = Heads**<br>
Your money will be 97, your bet amount will be 3<br>
**Round 5 = Heads**<br>
Your money will be 97, your bet amount will be 3<br>
<br>
and so on...

```python
#Initial Condition
Money=100
Bet=1
#Betting Strategy
Wallet=[]
while Money>0:
    Money=Money-Bet
    if coinflip()==1:
        Money=Money+Bet
    else:
        Bet+=1
    Wallet.append(Money)
print(Wallet)
```

Let's find out **when** will you run out of money. Well, the graph below shows that we will run out of money in 20 rounds, so... 25 rounds will be a pretty difficult challenge, right?

```python
plt.plot(Wallet)
plt.show()
```

## Random Walk

Let's simulate the random process above 100 times to see the possible number of total rounds until we run out of money.

```python
#Monte Carlo Simulation
All_Wallet=[]
for i in range(100):
    Money=100
    Bet=1
    Wallet=[]
    while Money>0:
        Money=Money-Bet
        if coinflip()==1:
            Money=Money+Bet
        else:
            Bet+=1
        Wallet.append(Money)
    All_Wallet.append(Wallet)
    
arr=np.transpose(np.array(All_Wallet))
```

```python
list(map(plt.plot,arr))
plt.show()
```

Apparently, the possible number of rounds varies! Let's draw a histogram to view the distribution of the total number of rounds.

```python
#list of possible total rounds
End=[]
for i in range(100):
    End.append(len(arr[i]))
print(End)
```

```python
#histogram
plt.hist(End)
plt.show()
```

```python
#Probability of having total rounds more than equal to 25 rounds
for i in End:
    if i<25:
        End.remove(i)
    else:
        next
len(End)/100
```

We can see clearly that we have a high probability (80%) to win the challenge and hold the money for more than 25 rounds!
